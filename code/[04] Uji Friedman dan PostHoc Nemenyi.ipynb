```python
import matplotlib.pyplot as plt
import numpy as np

# Data dari tabel
metrik = ['Akurasi', 'Recall', 'Precision', 'F1-Score']

# Skor masing-masing model
indobert = [0.9012, 0.8889, 0.8353, 0.8589]
naive_bayes = [0.6528, 0.6048, 0.5544, 0.5685]
svm = [0.7980, 0.6812, 0.7297, 0.7014]

# Pengaturan posisi diagram batang
x = np.arange(len(metrik))  # lokasi label
width = 0.25                # lebar setiap batang

# Membuat figure dan axis
fig, ax = plt.subplots(figsize=(10, 6))

# Membuat batang untuk masing-masing model
batang1 = ax.bar(x - width, indobert, width, label='IndoBERT', color='#2ca02c')
batang2 = ax.bar(x, naive_bayes, width, label='Naive Bayes', color='#ff7f0e')
batang3 = ax.bar(x + width, svm, width, label='SVM', color='#1f77b4')

# Menambahkan teks, label, judul, dan sumbu x khusus
ax.set_ylabel('Nilai Metrik Evaluasi', fontsize=12)
ax.set_title('Perbandingan Hasil Evaluasi Model IndoBERT, Naive Bayes, dan SVM', fontsize=12, fontweight='bold', pad=15)
ax.set_xticks(x)
ax.set_xticklabels(metrik, fontsize=11)
ax.set_ylim(0, 1.1) # Batas y-axis dilebihkan ke 1.1 agar ada ruang untuk teks nilai di atas batang

# Menambahkan legenda
ax.legend(loc='upper right', fontsize=11)

# Fungsi untuk menambahkan nilai angka di atas setiap batang
def tambah_label(batang_plot):
    for batang in batang_plot:
        tinggi = batang.get_height()
        ax.annotate(f'{tinggi:.4f}',
                    xy=(batang.get_x() + batang.get_width() / 2, tinggi),
                    xytext=(0, 3),  # 3 poin pergeseran vertikal
                    textcoords="offset points",
                    ha='center', va='bottom', fontsize=9)

# Menerapkan fungsi label ke ketiga model
tambah_label(batang1)
tambah_label(batang2)
tambah_label(batang3)

# Menampilkan grid horizontal tipis agar mudah dibaca
ax.yaxis.grid(True, linestyle='--', alpha=0.7)

# Merapikan layout dan menampilkan plot
fig.tight_layout()
plt.show()
```

```python
from scipy.stats import friedmanchisquare

# Data Macro F1-Score dari hasil 5-Fold CV
f1_indobert = [0.8756200662083015, 0.8509865890511777, 0.8274462990614296, 0.8470826428784918, 0.8589196991609612]
f1_nb = [0.5624495573623277, 0.5379469941074525, 0.5820201449527763, 0.5759025415466289, 0.5840108039946051]
f1_svm = [0.7117214760965395, 0.696447922249829, 0.6868218281121506, 0.6931256519506034, 0.7190872861929113]

# Uji Friedman
stat, p_value = friedmanchisquare(f1_indobert, f1_svm, f1_nb)

print("Hasil Uji Friedman")
print(f"P-Value: {p_value:.4f}")
```
